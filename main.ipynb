```python
# Function to calculate sub-index
def calculate_sub_index(Cp, Bp_lo, Bp_hi, I_lo, I_hi):
    return ((I_hi - I_lo)/(Bp_hi - Bp_lo)) * (Cp - Bp_lo) + I_lo


# PM2.5 sub-index
def aqi_pm25(cp):
    if cp <= 30:
        return calculate_sub_index(cp, 0, 30, 0, 50)
    elif cp <= 60:
        return calculate_sub_index(cp, 30, 60, 50, 100)
    elif cp <= 90:
        return calculate_sub_index(cp, 60, 90, 100, 200)
    elif cp <= 120:
        return calculate_sub_index(cp, 90, 120, 200, 300)
    elif cp <= 250:
        return calculate_sub_index(cp, 120, 250, 300, 400)
    else:
        return calculate_sub_index(cp, 250, 500, 400, 500)


# PM10 sub-index
def aqi_pm10(cp):
    if cp <= 50:
        return calculate_sub_index(cp, 0, 50, 0, 50)
    elif cp <= 100:
        return calculate_sub_index(cp, 50, 100, 50, 100)
    elif cp <= 250:
        return calculate_sub_index(cp, 100, 250, 100, 200)
    elif cp <= 350:
        return calculate_sub_index(cp, 250, 350, 200, 300)
    elif cp <= 430:
        return calculate_sub_index(cp, 350, 430, 300, 400)
    else:
        return calculate_sub_index(cp, 430, 600, 400, 500)


# NO2 sub-index
def aqi_no2(cp):
    if cp <= 40:
        return calculate_sub_index(cp, 0, 40, 0, 50)
    elif cp <= 80:
        return calculate_sub_index(cp, 40, 80, 50, 100)
    elif cp <= 180:
        return calculate_sub_index(cp, 80, 180, 100, 200)
    elif cp <= 280:
        return calculate_sub_index(cp, 180, 280, 200, 300)
    elif cp <= 400:
        return calculate_sub_index(cp, 280, 400, 300, 400)
    else:
        return calculate_sub_index(cp, 400, 500, 400, 500)


# SO2 sub-index
def aqi_so2(cp):
    if cp <= 40:
        return calculate_sub_index(cp, 0, 40, 0, 50)
    elif cp <= 80:
        return calculate_sub_index(cp, 40, 80, 50, 100)
    elif cp <= 380:
        return calculate_sub_index(cp, 80, 380, 100, 200)
    elif cp <= 800:
        return calculate_sub_index(cp, 380, 800, 200, 300)
    elif cp <= 1600:
        return calculate_sub_index(cp, 800, 1600, 300, 400)
    else:
        return calculate_sub_index(cp, 1600, 2000, 400, 500)


# CO sub-index (note: CO is in mg/m3)
def aqi_co(cp):
    if cp <= 1:
        return calculate_sub_index(cp, 0, 1, 0, 50)
    elif cp <= 2:
        return calculate_sub_index(cp, 1, 2, 50, 100)
    elif cp <= 10:
        return calculate_sub_index(cp, 2, 10, 100, 200)
    elif cp <= 17:
        return calculate_sub_index(cp, 10, 17, 200, 300)
    elif cp <= 34:
        return calculate_sub_index(cp, 17, 34, 300, 400)
    else:
        return calculate_sub_index(cp, 34, 50, 400, 500)


# O3 sub-index
def aqi_o3(cp):
    if cp <= 50:
        return calculate_sub_index(cp, 0, 50, 0, 50)
    elif cp <= 100:
        return calculate_sub_index(cp, 50, 100, 50, 100)
    elif cp <= 168:
        return calculate_sub_index(cp, 100, 168, 100, 200)
    elif cp <= 208:
        return calculate_sub_index(cp, 168, 208, 200, 300)
    elif cp <= 748:
        return calculate_sub_index(cp, 208, 748, 300, 400)
    else:
        return calculate_sub_index(cp, 748, 1000, 400, 500)
```

```python
def calculate_aqi(row):
    sub_indices = []

    sub_indices.append(aqi_pm25(row['PM2.5 (ug/m3)']))
    sub_indices.append(aqi_pm10(row['PM10 (ug/m3)']))
    sub_indices.append(aqi_no2(row['NO2 (ug/m3)']))
    sub_indices.append(aqi_so2(row['SO2 (ug/m3)']))
    sub_indices.append(aqi_co(row['CO (mg/m3)']))
    sub_indices.append(aqi_o3(row['Ozone (ug/m3)']))

    return max(sub_indices)


df_model['AQI'] = df_model.apply(calculate_aqi, axis=1) 
```

```python
def aqi_category(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

df_model['AQI_Category'] = df_model['AQI'].apply(aqi_category)

df_model[['AQI', 'AQI_Category']].head()

# AQI values were categorized into standard air quality levels based on CPCB classification to enable categorical analysis and visualization.
```

```python
import numpy as np

X_train = np.nan_to_num(X_train)
X_test = np.nan_to_num(X_test)

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Results:")
print("MAE:", mae_lr)
print("MSE:", mse_lr)
print("R2 Score:", r2_lr)



# Linear Regression was used as a baseline model to understand the linear relationship between pollutants and PM2.5 levels.

print()
"""-> The Linear Regression model achieved an R² score of approximately 0.79,
indicating a strong relationship between features and PM2.5. The MAE of around 30 suggests reasonable prediction 
accuracy given the variability in pollution data."""
```
